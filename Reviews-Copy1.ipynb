```python
from bs4 import BeautifulSoup
import requests

import re
import random
import time
```

```python
num_reviews = 1460
```

```python
url_list = []
for i in range(0,num_reviews,10):
    url_list.append('https://www.yelp.com/biz/the-palace-grill-santa-barbara?osq=The%20Palace%20Grill&start='+str(i))
```

```python
len(url_list)
```

```python
url_list
```

```python
allreviews = list()
revlen = []
for index, url in enumerate(url_list):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    allrev = soup.find_all('div', attrs={'class':' review__373c0__13kpL border-color--default__373c0__2oFDT'})
    u = list()
    for i in range(0,len(allrev)):
        l={}
        try:
            l["username"]=allrev[i].find("a",{'class':'css-166la90'}).text
        except:
            l["username"]=None
        try:
            l["location"]=allrev[i].find("span",{'class':' css-n6i4z7'}).text
        except:
            l["location"]=None
        try:
            l["date"]=allrev[i].find('span',{'class':' css-e81eai'}).text
        except:
            l["date"]=None
        try:
            l["stars"]=allrev[i].find("div",{"role":"img"}).get('aria-label')
        except:
            l["stars"]=None
        try:
            l["review"]=allrev[i].find("p",{'class':'comment__373c0__1M-px css-n6i4z7'}).text
        except:
            l["review"]=None
        u.append(l)
    revlen.append(len(allrev))
    allreviews = allreviews + u
    time.sleep(random.randint(10,20)) # Random sleep to avoid getting banned from the server
    print(len(allrev))
    print('Finished page' + str(index+1))
```
